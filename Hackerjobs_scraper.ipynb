```python
import pandas as pd
import requests
import re
response = requests.get('https://news.ycombinator.com/jobs')
print(response.status_code)
content=response.content
```

```python
from bs4 import BeautifulSoup
parser=BeautifulSoup(content,'html.parser')
table_row=parser.find_all('a')
table_row
```

```python
li=[]
for item in table_row:
    if len(item.text)>25:
        li.append([item.text,item['href']])
for i in li:
    if 'item?id' in i[1]:
        i[1]='https://news.ycombinator.com/'+i[1]
```

```python
jobs_title_data=pd.read_csv('/Users/Jaspinder/Desktop/five_referrals/job_table.csv')
existing_jobs=jobs_title_data['Job_Title'].tolist()
#### There was a bug because we were not using re. Some of the jobs like ones with '-' were getting append again and again
existing_jobs=[re.sub('[^A-Za-z1-9 ]','',i) for i in existing_jobs]
append_list=[]
for i in li:
    j=re.sub('[^A-Za-z1-9 ]','',i[0])
    if j not in existing_jobs:
        append_list.append(i)
for i in append_list:
    i.append('')
    i.append('')
append_list[:9]
```

```python
company_data=pd.read_csv('/Users/Jaspinder/Desktop/five_referrals/customer_table.csv')
li_co_names=company_data['Name'].tolist()
li_co_description=company_data['Description'].tolist()
for i in li_co_names:
    for j in append_list:
        if i.lower() in j[0].lower():
            j[2]=i
            j[3]= li_co_description[li_co_names.index(i)]
append_list
```

```python
cols=['Job_Title', 'JD_Link', 'Company_Name', 'Company_Description']
new_jobs_df=pd.DataFrame(append_list,columns=cols)
new_jobs_df.head()
```

```python
jobs_title_data=jobs_title_data.append(new_jobs_df)
jobs_title_data=jobs_title_data[['Job_Title', 'JD_Link', 'Company_Name', 'Company_Description']]
jobs_title_data
```
